2.Is there a significant difference in salary based on different job designations?

```python
top_designations = data['Designation'].str.strip().value_counts().nlargest(10).index

data_filtered = data[data['Designation'].str.strip().isin(top_designations)]

plt.figure(figsize=(5, 5))
sns.barplot(x='Designation', y='Salary', data=data_filtered, order=top_designations)
plt.title('Top 10 Designations vs Salary')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Designation')
plt.ylabel('Salary')
plt.show()
```

```python
# Designation vs. Salary 
designations = data['Designation'].unique()
designation_salary_list = [data['Salary'][data['Designation'] == designation] for designation in designations]

anova_result = f_oneway(*designation_salary_list)

print("ANOVA Result - Designation vs. Salary:")
print(anova_result)
```

```python

mean_salary_by_designation = data.groupby('Designation')['Salary'].mean()

print("Mean Salary by Designation:")
print(mean_salary_by_designation)

```

OBSERVATIONS: The mean salaries vary across different designations. 1.Some designations, like 'account manager' and '.net developer', have relatively high mean salaries, suggesting that individuals in these roles earn more on average. 2.certain designations, such as 'admin assistant' or 'web intern', have lower mean salaries, indicating that individuals in these roles earn less on average.

Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(data['Gender'], data['Specialization'])

# Chi-Square test
chi2, p, _, _ = chi2_contingency(contingency_table)

if p < 0.05:
    print("There is a significant relationship between gender and specialization.")
else:
    print("There is no significant relationship between gender and specialization.")
```
